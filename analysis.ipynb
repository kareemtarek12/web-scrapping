```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import re
```

```python
from bs4 import BeautifulSoup  as bsP
```

```python
from urllib.request import urlopen 
```

```python
url='https://forasna.com/a/%D9%88%D8%B8%D8%A7%D8%A6%D9%81-%D8%AA%D8%B3%D9%88%D9%8A%D9%82-%D9%88%D9%85%D8%A8%D9%8A%D8%B9%D8%A7%D8%AA-%D9%81%D9%8A-%D9%85%D8%B5%D8%B1'
```

```python
client=urlopen(url)
```

```python
HTML=client.read()
HTML
```

```python
client.close()
```

```python
soup=bsP(HTML,'html.parser')
soup
```

```python
container=soup.find_all("div",{"class":"result-wrp"})
```

```python
extradetails=container[12].findAll("span",{"class":"job-details__title"})
ex_cleaned=extradetails[1].find_next("span").text.strip()

ex_cleaned=re.sub(r'\s+',"",ex_cleaned)
ex_cleaned
```

```python
for i in container:
    jobtitle=i.findAll('h2',{"class":"job-title"})
    jtitle=jobtitle[0].text.strip()

    cname=i.findAll("span",{"class":"company-name"})
    company_name=cname[0].text.strip()
    
    loc=i.findAll("span",{"class":"location location-desktop"})
    location=loc[0].text.strip() 

   

    extradetails = i.findAll("span", {"class": "job-details__title"})

    try:
         ex_cleaned = extradetails[1].find_next("span").text.strip()
         ex_cleaned = re.sub(r'\s+', "", ex_cleaned)
    except IndexError:
         ex_cleaned = "No extra details available"  
    print(jtitle+","+company_name+","+location+","+ex_cleaned+"\n\n")
    
```
